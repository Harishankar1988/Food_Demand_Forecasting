```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor 
```

```python
df_train = pd.read_csv('input/train_processed.csv')
df_submit = pd.read_csv('input/test_processed.csv')
```

```python
#Preparing dtaframe for catboost
df_train = df_train.drop(['Unnamed: 0','id','center_id', 'meal_id',
       'city_code', 'region_code', 'city_region',
       'center_type', 'category', 'cuisine','num_orders',
        'city_code', 'region_code'], axis=1)
df_tmp = df_submit.drop(['Unnamed: 0','id','center_id', 'meal_id',
       'city_code', 'region_code', 'city_region',
       'center_type', 'category', 'cuisine',
        'city_code', 'region_code'], axis=1)
```

```python
df_train['Quarter']=(df_train['week']/13).astype('int64')
df_train['Quarter'] = df_train['Quarter'].map({0:'Q1',
                         1:'Q2',
                         2:'Q3',
                         3:'Q4',
                         4:'Q1',
                         5:'Q2',
                         6:'Q3',
                         7:'Q4',
                         8:'Q1',
                         9:'Q2',
                         10:'Q3',
                         11:'Q4'})

df_train['Year']=(df_train['week']/52).astype('int64')
df_train['Year'] = df_train['Year'].map({0:'Y1',
                         1:'Y2',
                         2:'Y3',
                         3:'Y4',
                         4:'Y5'})

df_tmp['Quarter']=(df_tmp['week']/13).astype('int64')
df_tmp['Quarter'] = df_tmp['Quarter'].map({0:'Q1',
                         1:'Q2',
                         2:'Q3',
                         3:'Q4',
                         4:'Q1',
                         5:'Q2',
                         6:'Q3',
                         7:'Q4',
                         8:'Q1',
                         9:'Q2',
                         10:'Q3',
                         11:'Q4'})

df_tmp['Year']=(df_tmp['week']/52).astype('int64')
df_tmp['Year'] = df_tmp['Year'].map({0:'Y1',
                         1:'Y2',
                         2:'Y3',
                         3:'Y4',
                         4:'Y5'})
```

```python
cat_cols_catboost = ['week','meal_id_encoded',
                     'center_id_encoded',
                     'emailer_for_promotion',
                     'homepage_featured',
                     'city_region_encoded',
                     'center_type_encoded',
                     'category_encoded',
                     'cuisine_encoded',
                    'Quarter',
                    'Year'
                    ]
```

```python
df_trn = df_train[df_train['week'] < 135]
df_tst = df_train[df_train['week'] > 134]
```

```python
y_train = df_trn['num_orders_log1p']
X_train = df_trn.drop('num_orders_log1p', axis=1)

y_test = df_tst['num_orders_log1p']
X_test = df_tst.drop('num_orders_log1p', axis=1)
```

```python
model = CatBoostRegressor(cat_features = cat_cols_catboost, ignored_features = ['week_cos','discount','base_price'], early_stopping_rounds=500)
```

```python
#grid = {'learning_rate': [0.01, 0.1, 1], 'depth':[4,6,8]}
```

```python
#grid_search_resut = model.grid_search(grid, X=X_train, y=y_train, plot=True)
```

```python
#params = grid_search_resut.get('params')
```

```python
model.fit(X_train, y_train, cat_features = cat_cols_catboost)
```

```python
y_predict = model.predict(X_test)
y_predict = np.expm1(y_predict)
```

```python
#y_pred = np.expm1(y_predict)
```

```python
plt.scatter(y_test, y_predict)
```

```python
y_pred_s = pd.Series(y_predict)
count, division = np.histogram(y_pred_s)
print(f"Data_Slice: {division}")
print(f"Count: {count}")
y_pred_s.describe()
```

```python
y_pred_less_than_0 = y_pred_s[y_pred_s < 0]
print(y_pred_less_than_0.values)
```

```python
# negative requirement doesn't mean anything.
y_predict[y_predict < 0] = 0
```

```python
plt.scatter(y_test, y_predict)
```

```python
from sklearn.metrics import mean_squared_log_error
np.sqrt(mean_squared_log_error( y_test, y_predict ))
```
